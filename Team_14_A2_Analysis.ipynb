# Loading Libraries & Functions

```python
########################################
# importing packages
########################################
import warnings
warnings.filterwarnings('ignore')

import numpy as np                                      # mathematical essentials
import pandas            as pd                          # data science essentials
import matplotlib.pyplot as plt                         # fundamental data visualization
import seaborn           as sns                         # enhanced visualizations
from sklearn.preprocessing import StandardScaler        # standard scaler
from sklearn.decomposition import PCA                   # pca
from scipy.cluster.hierarchy import dendrogram, linkage # dendrograms
from sklearn.cluster         import KMeans              # k-means clustering


##########################################
# Creating a User-Defined function: inertia
##########################################
def inertia_plot(data, max_clust = 50):
    """
PARAMETERS
----------
data      : DataFrame, data from which to build clusters. Dataset should be scaled
max_clust : int, maximum of range for how many clusters to check interia, default 50
    """
    ks = range(1, max_clust)
    inertias = []

    for k in ks:
        # INSTANTIATING a kmeans object
        model = KMeans(n_clusters = k)

        # FITTING to the data
        model.fit(data)

        # append each inertia to the list of inertias
        inertias.append(model.inertia_)

    # plotting ks vs inertias
    fig, ax = plt.subplots(figsize = (12, 8))
    plt.plot(ks, inertias, '-o')

    # labeling and displaying the plot
    plt.xlabel('number of clusters, k')
    plt.ylabel('inertia')
    plt.xticks(ks)
    plt.show()

    
############################################
# Defining function for categorical Boxplots
############################################
def categorical_boxplots(y, x, data):
    """
	This function is designed to generate a boxplot for  can be used for categorical variables.
    Make sure matplotlib.pyplot and seaborn have been imported (as plt and sns).

    PARAMETERS
	----------
	y : str, response variable
	x  : str, categorical variable
	data     : DataFrame of the response and categorical variables
	"""

    fig, ax = plt.subplots(figsize = (12, 12))
    
    sns.boxplot(x    = x,
                y    = y,
                hue  = 'Cluster',
                data = data)
    
    plt.suptitle("")
    plt.show()
    
    
    
########################################
# scree_plot
########################################
def scree_plot(pca_object, export = False):

    fig, ax = plt.subplots(figsize=(10, 8))
    features = range(pca_object.n_components_)

    # developing a scree plot
    plt.plot(features,
             pca_object.explained_variance_ratio_,
             linewidth = 2,
             marker = 'o',
             markersize = 10,
             markeredgecolor = 'black',
             markerfacecolor = 'grey')

    # setting more plot options
    plt.title('Scree Plot')
    plt.xlabel('PCA feature')
    plt.ylabel('Explained Variance')
    plt.xticks(features)

    if export == True:

        plt.savefig('./analysis_images/top_customers_correlation_scree_plot.png')

    plt.show()
```

### Big 5 Preprocessing
<hr style="height:.9px;border:none;color:#333;background-color:#333;" />

Once we subset our data into our 3 distinct assesments , we focus now on preprocessing **Big 5** section.

Initially we will:

<h4>• Reverse negative questions: </h4>

By reversing negative questions, we are putting our scores on the same scale where we will be able to measure
the presence or lack of a specific personality trait. We will remove the original negatives ones.

<h4>• Individual trait score: </h4>

Add all the scores per personality trait <br>


<h4>• Check for contradicting answers: </h4>

In order to have a more accurate assesment and subsequent analysis, we decided to look at two specific pair of questions that are stated as the exact opposite of the other.

--> "I am interested in people"  --   "Am not really interested in others"

--> "Often feel blue"   ---  'Seldom feel blue"

Respondents who answered any of these pair of questions with the same score (1-1) or (5-5) give us a hint as to the unreliability of their answers in the survey. These respondents will be removed.


<h4>• Remove unreliable respondents: </h4>

Remove the respondents that contradicted themselves on very clear opposite questions. REspondents should be removed from the whole dataset

<br><br>


```python
##################################################
# Creating dataframes with all questions per trait
##################################################

extraversion        = survey_df[big_five['Extraversion']]
agreeableness       = survey_df[big_five['Agreeableness']]
conscientiousness   = survey_df[big_five['Conscientiousness']]
neuroticism         = survey_df[big_five['Neuroticism']]
openness            = survey_df[big_five['Openness']]


###############################################
# Reversing negative questions
###############################################

reversed_questions = survey_df[big_five['reverse']]

# Loop to reverse the score of negative questions
for col in reversed_questions:
    
    reversed_questions["rev_"+col] = 6 - reversed_questions.loc[:, col]

# Keep only reversed questions
reversed_questions = reversed_questions.drop(big_five['reverse'], axis=1)


######################################
# Checking for contradicting answers
######################################

index_list=[]

for index, col in agreeableness.iterrows():
    if ((agreeableness.loc[index, 'Am interested in people']==1) & (agreeableness.loc[index, 'Am not really interested in others']==1)):
        index_list.append(index)
    elif ((agreeableness.loc[index, "Am interested in people"]==5) & (agreeableness.loc[index, "Am not really interested in others"]==5)):
        index_list.append(index)
    else:
        continue
        
for index, col in neuroticism.iterrows():
    if ((neuroticism.loc[index, 'Seldom feel blue']==1) & (neuroticism.loc[index, 'Often feel blue']==1)):
        index_list.append(index)
    elif ((neuroticism.loc[index, "Seldom feel blue"]==5) & (neuroticism.loc[index, "Often feel blue"]==5)):
        index_list.append(index)
    else:
        continue

index_list.sort()


#################################################
#Dropping the negative questions, replacing them with reversed
#################################################

extraversion =      extraversion.drop(["Don't talk a lot",
                                       "Keep in the background",
                                       "Have little to say",
                                       "Don't like to draw attention to myself",
                                       "Am quiet around strangers"], axis=1)

agreeableness =     agreeableness.drop(["Am not really interested in others",
                                       "Insult people",
                                       "Am not interested in other people's problems",
                                       "Feel little concern for others"], axis=1)

conscientiousness = conscientiousness.drop(["Leave my belongings around",
                                            "Make a mess of things",
                                            "Often forget to put things back in their proper place",
                                            "Shirk my duties"], axis=1)

neuroticism =       neuroticism.drop(["Am relaxed most of the time","Seldom feel blue"], axis=1)

openness =          openness.drop(["Have difficulty understanding abstract ideas",
                                   "Am not interested in abstract ideas",
                                   "Do not have a good imagination"], axis=1)


#Adding the reversed questions
extraversion = extraversion.join(reversed_questions[["rev_Don't talk a lot",
                                      "rev_Keep in the background",
                                      "rev_Have little to say",
                                      "rev_Don't like to draw attention to myself"]])

agreeableness = agreeableness.join(reversed_questions[["rev_Am not really interested in others",
                                                       "rev_Insult people",
                                                       "rev_Am not interested in other people's problems",
                                                       "rev_Feel little concern for others"]])

conscientiousness = conscientiousness.join(reversed_questions[["rev_Leave my belongings around",
                                                               "rev_Make a mess of things",
                                                               "rev_Often forget to put things back in their proper place",
                                                               "rev_Shirk my duties"]])

neuroticism = neuroticism.join(reversed_questions[["rev_Am relaxed most of the time", "rev_Seldom feel blue"]])


openness = openness.join(reversed_questions[["rev_Have difficulty understanding abstract ideas",
                                             "rev_Am not interested in abstract ideas",
                                             "rev_Do not have a good imagination"]])

#########################
# Scoring per personality
#########################

extraversion["total_ext"]       = extraversion.sum(axis=1)
agreeableness["total_agree"]    = agreeableness.sum(axis=1)
conscientiousness["total_cons"] = conscientiousness.sum(axis=1)
neuroticism["total_neur"]       = neuroticism.sum(axis=1)
openness["total_open"]          = openness.sum(axis=1)


######################################
# Joining scores in a single dataframe
######################################

personalities_df = pd.DataFrame({'EXTR': extraversion["total_ext"],
                      'AGREE': agreeableness["total_agree"],
                      'CONS':conscientiousness["total_cons"],
                      'NEUR': neuroticism["total_neur"],
                      'OPEN': openness["total_open"]})


########################################
# dropping index rows of bad respondents
########################################

personalities_df = personalities_df.drop(index_list, axis=0)
```

### Hult DNA Questions Preprocessing
<hr style="height:.9px;border:none;color:#333;background-color:#333;" />

We will now go through a very similar preprocessing step for the **Hult DNA** section.

Initially we will:

<h4>• Reverse negative questions: </h4>

Putting scores on the same scale where we will be able to measure the presence or lack of a specific personality trait. 

<h4>• Subset: </h4>

Group all questions that concern to a specific Hult DNA dimension <br>


<h4>• Dimension score: </h4>

Add all the scores per Hult DNA dimension.


<h4>• Remove unreliable respondents: </h4>

Remove the respondents that contradicted themselves on very clear opposite questions. Respondents should be removed from the whole dataset

<br><br>


```python
###################################################
# Reversing negative questions & dropping negatives
###################################################

reverse_columns_hult = ["Don't  generate ideas that are new and different", 
                        "Don't persuasively sell a vision or idea", 
                        "Can't rally people on the team around a common goal"]

for col in reverse_columns_hult:
    
    hultDNA_df['reverse_'+col] = [6-x for x in hultDNA_df[col]]


hultDNA_df   = hultDNA_df.drop(["Don't  generate ideas that are new and different", 
                                "Don't persuasively sell a vision or idea", 
                                "Can't rally people on the team around a common goal"],
                                axis = 1)


##################
# Subsetting
###################

thinking = hultDNA_df.loc[:, ["See underlying patterns in complex situations", 
                              "Demonstrate an awareness of personal strengths and limitations",
                              "Display a growth mindset", "Seek and use feedback from teammates",
                              "reverse_Don't  generate ideas that are new and different"]]

communicating = hultDNA_df.loc[:, ["Listen carefully to others",
                                   "Effectively negotiate interests, resources, and roles", 
                                   "Translate ideas into plans that are organized and realistic", 
                                   "reverse_Don't persuasively sell a vision or idea",
                                   "Respond effectively to multiple priorities", 
                                   "Take initiative even when circumstances, objectives, or rules aren't clear", 
                                   "Drive for results"]]

team = hultDNA_df.loc[:, ["Encourage direct and open discussions", "Build cooperative relationships", 
                          "Work well with people from diverse cultural backgrounds", 
                          "Resolve conflicts constructively", "Coach teammates for performance and growth", 
                          "reverse_Can't rally people on the team around a common goal"]]


################################
# Scoring per Hult Dna Dimension
################################


hultDNA_df['thinking_dna']      = thinking.sum(axis=1)
hultDNA_df['communicating_dna'] = communicating.sum(axis=1)
hultDNA_df['team_dna']          = team.sum(axis=1)


hultDNA_df_sums = hultDNA_df.loc[:,['thinking_dna', 'communicating_dna', 'team_dna']]


########################################
# dropping index rows of bad respondents
########################################

hultDNA_df_sums = hultDNA_df_sums.drop(index_list, axis=0)
```

### Demographics Preprocessing
<hr style="height:.9px;border:none;color:#333;background-color:#333;" />

The cleaning process for the **Demographics** section will be slightly different from the previous two since we are dealing with categorical data:

We will proceeed as follows:

<h4>• Sorting nationalities into regions of the world: </h4>

Group respondents nationalities into geographical regions: North America, South America, Africa, Europe, Asia and undefined.

<h4>• Subset: </h4>

Group all questions that concern to a specific Hult DNA dimension <br>


<h4>• Score: </h4>

Add all the scores per Hult DNA dimension


<h4>• Remove unreliable respondents: </h4>

Remove the respondents that contradicted themselves on very clear opposite questions. REspondents should be removed from the whole dataset

<br><br>

```python
####################################
# Regions Sorting
####################################

africa        = ['Congolese (Dr Congo)','Ugandan','Nigerian', 'Mauritius',
                'Congolese','Nigeria', 'Kenyan', 'Ghanaian','Congolese (Dr Congo)']

north_america = ['Usa','American','Canada', 'Canadian' ]

south_america = ['Ecuador','Dominican ','Costarrican','Mexican','Peruvian','Peru',
                 'Colombian', 'Brazil','Brazilian', 'Colombia','Venezuelan','Panama',
                 'Dominican','Ecuadorian' ]

europe        = ['German/American', 'British, Indian', 'Swiss', 'German', 'Italian', 'Norwegian','Italian And Spanish', 
                 'Czech Republic', 'Belarus','Czech','Ukrainian','Belgian ', 'Germany',
                 'British', 'Portuguese','Spain','Spanish' , 'Russian','Russia' ]
undefined     = ['Prefer Not To Answer',]

asia          = [ 'China','Japan','Chinese','Filipino ','Philippines','Indonesia', 
                 'Thai','Turkish', 'South Korea', 'Korea','Vietnamese',
                 'Indonesian','Taiwan',  'Kyrgyz', 
                 'Republic Of Korea', 'Pakistani', 'Indian.','Indian']


# Create a new column using .str.title() for better string manipulation
demographics_df["nationality_new"]= demographics_df['What is your nationality? '].str.title()


# Region Loop
placeholder_lst = []

# creating a loop to populate a list with the regions
for nationality in demographics_df['nationality_new']: 
    
    # looking for nationalities belonging to south america group
    if nationality in south_america:
        placeholder_lst.append('South America')
        
    # looking for nationalities belonging to Europe group
    elif nationality in europe: 
        placeholder_lst.append('Europe')
    
    # looking for nationalities belonging to Africa group    
    elif nationality in africa: 
        placeholder_lst.append('Africa')
    
    # looking for nationalities belonging to North America group    
    elif nationality in north_america: 
        placeholder_lst.append('North America')
   
    # looking for nationalities belonging to Unknown group    
    elif nationality in undefined: 
        placeholder_lst.append("NaN")
        
    # looking for nationalities belonging to Asia group    
    elif nationality in asia: 
        placeholder_lst.append('Asia')
    else: 
        print("something went wrong")

#Joining region to demographics DF
demographics_df["Region"] = pd.Series(placeholder_lst) 



###################
# Age Sorting
###################

# Sorting age in 3 groups.

twenty_one_25 = [21, 22, 23, 24, 25]
twenty_six_28 = [ 26, 27, 28]
twenty_nine_plus = [29, 30, 31, 32,33,34,35,36,38,40,41,43,45,48,49]

# creating a placeholder list for the loop to fill
placeholder_lst2 = []

for age in demographics_df["What is your age?"]:
    if age in twenty_one_25: 
        placeholder_lst2.append("Age: 21 to 25")
    elif age in twenty_six_28: 
        placeholder_lst2.append("Age: 26 to 28")
    elif age in twenty_nine_plus: 
        placeholder_lst2.append("Age: 29 to 49")
    else: 
        print("something went wrong")

# converting the placeholder list to a series and a column in our dataframe
demographics_df["Age_Group"] = pd.Series(placeholder_lst2) 



#######################
# Ethnic groups sorting
#######################

minority_respondents = ["African American","Middle Eastern",
                        "Native American","Native American",
                        "Prefer not to answer"]


# creating a placeholder list for the loop to fill
placeholder_lst2 = []

for ethnicity in demographics_df["What is your ethnicity?"]:
    if ethnicity in minority_respondents: 
        placeholder_lst2.append("Minority Respondents")
    elif ethnicity not in minority_respondents:
        placeholder_lst2.append(ethnicity)
    else: 
        print("something went wrong") 

# converting the placeholder list to a series and a column in our dataframe
demographics_df["Ethnicity_New"] = pd.Series(placeholder_lst2)   



#######################################################################
# Renaming columns, subsetting and dropping index rows (bad respondents)
#######################################################################

# renaming demographics columns 
demographics_df["Laptop_now"] = demographics_df["What laptop do you currently have?"]
demographics_df["Laptop_future"] = demographics_df["What laptop would you buy in next assuming if all laptops cost the same?"]
demographics_df["Program_Hult"] =demographics_df["What program are you in?"]

# dictionary for the clean demographics columns
dem_dict = { 'clean'  : ["Laptop_now", "Laptop_future", "Program_Hult", "Gender", 
              "Ethnicity_New", "Age_Group", "Region"]}

# creating a demographics dataframe with tidy column names and content. 
demographics_clean = demographics_df.loc[: , dem_dict["clean"]]


# dropping index rows of bad respondents
demographics_clean=demographics_clean.drop(index_list, axis=0)



###############################
# Reseting demographics indices
################################

# reindexing because when we join with scaled DFs later - they are re-indexed too
demographics_clean = demographics_clean.reset_index()
demographics_clean = demographics_clean.drop("index", axis=1)
```

# Joined Clustering
<hr style="height:.9px;border:none;color:#333;background-color:#333;" />

After loading each persona from the 2 separate PCA analysis (Big 5 and Hult DNA) into each of our respondents, we can now perform a cluster analysis to separate our respondents into groups with certain similarities and look for valuable insights for marketing purposes. <br><br>

## Instantiating clustering

```python
#######################################################################
# Saving the personas and factor loadings for each respondent Hult Dna
#######################################################################

# converting into a DataFrame 
hultDNA_pca_df = pd.DataFrame(hultDNA_pca)

# renaming columns according to personas from above.
hultDNA_pca_df.columns = factor_loadings_hult.columns



#######################################################################
# Saving the personas and factor loadings for each respondent  Big 5
#######################################################################

# converting into a DataFrame 
personalities_pca_df = pd.DataFrame(pca_personalities_fit)

# renaming columns according to personas from above.
personalities_pca_df.columns = factor_loadings_personalities.columns



###############################################
# Merging the two loadings for cluster analysis
###############################################

joined_pca_factor_loadings = personalities_pca_df.join(hultDNA_pca_df)

#print(joined_pca_factor_loadings)



###############################################
# Scaling our data Before instantiating K means
###############################################

scaler = StandardScaler()

scaler.fit(joined_pca_factor_loadings)

joined_pca_fl_scaled = scaler.transform(joined_pca_factor_loadings)

joined_pca_fl_scaled_df = pd.DataFrame(joined_pca_fl_scaled)

# reattaching column names
joined_pca_fl_scaled_df.columns = joined_pca_factor_loadings.columns



#########################################
# Test the number of cluster with k-Means
#########################################

# Instantiating k-Means object with 3 clusters
respondents_k_joined = KMeans(n_clusters   = 3,
                         random_state   = 219)

# fitting the object to the data
respondents_k_joined.fit(joined_pca_fl_scaled_df)

# converting the clusters to a DataFrame
respondents_kmeans_joined_df = pd.DataFrame({'Cluster': respondents_k_joined.labels_})

# checking the results
print(f"""

Cluster sizes:

{respondents_kmeans_joined_df.iloc[: , 0].value_counts(sort=False)}""")




###################################
# Identifying the cluster centroids
###################################

# storing cluster centers
centroids_joined = respondents_k_joined.cluster_centers_

# converting cluster centers into a DataFrame
cluster_joined_df = pd.DataFrame(centroids_joined)

# renaming principal components
cluster_joined_df.columns = ["Goody_Goody", 
                             "Introverted_Neat_Freak", 
                             "Stressed_Out_Friend", 
                             "Creative_Genius", 
                             'Non_Hultian', 
                             'Thoughtless_Leader']

# checking results (clusters = rows, pc = columns)
cluster_joined_df.round(2)
```

## Joining demographics and cluster results
<hr style="height:.9px;border:none;color:#333;background-color:#333;" />


 <br><br>

```python
##########################################
# Joining demographics and cluster results
##########################################

# concatenating cluster memberships with principal components
pca_cluster_df = pd.concat([respondents_kmeans_joined_df,
                         joined_pca_fl_scaled_df],
                         axis = 1)

# concatenating demographic information with pca-clusters
pca_cluster_df = pd.concat([demographics_clean.loc[ : , :],
                                pca_cluster_df],
                                axis = 1)

# renaming columns
pca_cluster_df.columns = ["Laptop_now", 
                            "Laptop_future",
                            "Program_Hult",
                            "Gender", 
                            "Ethnicity_New", 
                            "Age_Group", 
                            "Region",
                            "Loyal_MacBook", 
                            "Switch_to_MacBook",
                            'Cluster',
                            "Goody_Goody", 
                            "Introverted_Neat_Freak", 
                            "Stressed_Out_Friend", 
                            "Creative_Genius", 
                            'Non_Hultian', 
                            'Thoughtless_Leader']

# renaming regions
cluster_names = {0.0 : 'Cluster 0',
                 1.0 : 'Cluster 1', 
                 2.0 : 'Cluster 2'}


pca_cluster_df['Cluster'].replace(cluster_names, inplace = True)

# adding a productivity step
combined_final_df = pca_cluster_df
```

The crux about this group is that over 13.5% of them indicated that they would consider a switch to a MacBook in the future, provided that pricing is not part of the issue. This would mean a 25% growth in the proportion of Mac owners inside this&nbsp;subsection of people.</p>
<p style='margin-right:0cm;margin-left:0cm;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:18.0pt;color:#595959;text-align:start;'><span style='font-family:"Avenir Next",sans-serif;color:#595959;'>The opportunity and challenge lie where Apple needs to design a solution that appeals to generally change-averse people that are currently not being served by their Windows PC. The opportunity lies in the fact that this group of people is data-driven, logical as well as outgoing.</span></p>
<h2 style='margin-top:8.0pt;margin-right:0cm;margin-bottom:10.0pt;margin-left:0cm;text-align:start;font-size:27px;font-family:"Avenir Next",sans-serif;color:#0072C6;font-weight:bold;'>Cluster Analysis</h2>
<p style='margin-right:0cm;margin-left:0cm;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:150%;color:#595959;text-align:start;'>After using the PCA to reduce the complexity in the information contained within the Big 5 Personality Traits and Hult DNA survey portions, we applied the k means algorithm to identify patterns across the holistic view of a customer. The clusters that emerged provide a diverse target group with aspects from both surveys. After initial analysis, Cluster 0 emerged as the ideal target for Apple&rsquo;s future marketing efforts. It contains one-third of the total population and 10% of the people in this cluster would switch from Windows to MacBook the next time they buy a laptop.&nbsp;</p>
<p style='margin-right:0cm;margin-left:0cm;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:150%;color:#595959;text-align:start;'><span style="color:#3A3A3A;"><span style="color:windowtext;">People in the target cluster demonstrate self-awareness, high emotional intelligence, and resilience. They are good listeners, articulate speakers, and are able to present ideas effectively to persuade others. In addition to being outgoing, empathetic, and trustful, they are organized and hard-working. Similar to the personality analysis, this cluster is also&nbsp;</span></span><span style="color:#3A3A3A;"><span style="color:windowtext;">made up of a&nbsp;</span></span><span style="color:#3A3A3A;"><span style="color:windowtext;">majority of people (75%) pursuing a second master&rsquo;s degree in Business Analytics.</span></span></p>
<p style='margin-right:0cm;margin-left:0cm;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:150%;color:#595959;text-align:start;'><span style="color:#595959;">Overall, there is significant overlap between the personality analysis and the cluster, but the cluster provides a bit of a&nbsp;</span>broader view and target group. The following actionable insights are designed to appeal to both the persona specifically and the group of people in cluster 0 as a whole.<span style="color:#595959;">&nbsp;</span></p>
<h2 style='margin-top:8.0pt;margin-right:0cm;margin-bottom:10.0pt;margin-left:0cm;text-align:start;font-size:27px;font-family:"Avenir Next",sans-serif;color:#0072C6;font-weight:bold;'>Actionable Suggestions:</h2>
<p style='font-size:16px;font-family: "Avenir Next", sans-serif;color: rgb(89, 89, 89);margin: 8pt 0cm 16pt;line-height: 18pt;text-align: center;'><span style='font-family:"Avenir Next",sans-serif;color:#595959;'>&ldquo;Once you switch, you will never have to switch again&rdquo;</span></p>
<p style='margin-right:0cm;margin-left:0cm;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:150%;color:#595959;text-align:start;'>The Apple&reg; ecosystem encompasses many devices that interact seamlessly with each other. A business analytics student can take notes, extend her screen or plan her day using any combination of iPhone&reg;, iPad&reg;, MacBook&reg; or even the Apple Watch&reg;. The full features can be used even better when Apple iCloud&reg; is used to synchronize information across these devices and make for a familiar experience &ndash; on all devices, present, and future. Once a customer has bought into the Apple &reg; system, any changes in the future will become painless, and adding another device will be smooth as well,&nbsp;because iCloud enables this easy experience.&nbsp;</p>
<p style='margin-right:0cm;margin-left:18.0pt;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:18.0pt;color:#595959;text-align:start;'><span style='font-family:"Avenir Next",sans-serif;color:#595959;'>&nbsp;</span></p>
<div style='margin-top:8.0pt;margin-right:0cm;margin-bottom:16.0pt;margin-left:0cm;line-height:150%;font-size:16px;font-family:"Avenir Next",sans-serif;color:#595959;'>
    <p style="margin-left: 40px;">1. In order to fully demonstrate the simplicity and appeal to the fact-driven personas, Apple &reg; should offer a free one-year iCloud membership to new customers, that is big enough (25-50GB) to allow for photos, documents, contacts, and settings to sync across all potential devices. </p>
</div>
<p style='margin-right:0cm;margin-left:18.0pt;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:18.0pt;color:#595959;text-align:start;'><span style='font-family:"Avenir Next",sans-serif;color:#595959;'>This will be a direct value-added to our persona, as information can be shared between devices and it invites these customers to take advantage of the whole Apple &reg; Ecosystem. Once in the system, the new customers can rest easy in the familiarity and simplicity of the system.&nbsp;</span></p>
<p style='margin-right:0cm;margin-left:18.0pt;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:18.0pt;color:#595959;text-align:start;'><span style='font-family:"Avenir Next",sans-serif;color:#595959;'>To the people in the cluster, this offer is also quite an incentive, as they appreciate the order and structure that comes with this synchronicity (Rivera, 2019). Having an iCloud will also allow for collaboration, document sharing, and presenting with their teammates in university and professionally in the future.&nbsp;</span></p>
<div style='margin-top:8.0pt;margin-right:0cm;margin-bottom:16.0pt;margin-left:0cm;line-height:150%;font-size:16px;font-family:"Avenir Next",sans-serif;color:#595959;'>
    <p style="margin-left: 60px;">2. To appeal to the social and outgoing nature of our target groups Apple &reg; can offer a special Refer-a-Friend promotion for the iCloud subscription. For every person the customer invites that also subscribes to iCloud, her subscription gets extended by one month.</p>
</div>
<p style='margin-right:0cm;margin-left:0cm;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:150%;color:#595959;text-align:start;'>This allows their social side to shine through, and at the same time, it pulls at their pragmatism, that they gain a free month of iCloud for themselves. Oftentimes, a recommendation is more a way for the customer herself to receive recognition and strengthen her sense of belonging. (Cockburn, 2020) In that sense, this scheme is a great method to create a network of subscribers to the Apple &reg; cloud service that is compatible with both Mac and Microsoft laptop users. Ultimately, the best outcome will be to on-board Microsoft users to dip their toes into the Apple &reg; ecosystem in preparation for converting them, the next time they purchase a new computer.&nbsp;</p>
<h2 style='margin-top:8.0pt;margin-right:0cm;margin-bottom:10.0pt;margin-left:0cm;text-align:start;font-size:27px;font-family:"Avenir Next",sans-serif;color:#0072C6;font-weight:bold;'>Conclusion</h2>
<p style='margin-right:0cm;margin-left:0cm;font-size:16px;font-family:"Avenir Next",sans-serif;margin-top:8.0pt;margin-bottom:16.0pt;line-height:150%;color:#595959;text-align:start;'>Overall, this approach targets practical, organized people who also like to keep things simple. The Apple &reg; products are designed to be easy to use and integrate with each other. The glue that holds all that simplicity and order together is the Apple iCloud&reg; platform. Offering an extended free trial period with a newly purchased device and marketing it as a time-saving, optimization tool that simply runs in the background will appeal to both the persona we identified in the personalities and cluster 0 as a whole. Finally, the social characters of our target come into play when trying to convert change-averse, Windows users. Onboarding them via a referral program to the cloud service will let them get a taste of the simplicity and lets them feel connected with the wider audience for teamwork, file sharing, and calendar planning, even if they are starting out on a Windows&reg; laptop. This minimally invasive approach gives these users time to get to know the brand and takes advantage of their characteristics in the best possible way, to fully convert them in the long-run.</p>
